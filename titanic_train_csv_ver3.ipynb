```python
le = preprocessing.LabelEncoder()
data['Sex'] = le.fit_transform(data['Sex'])
data['Embarked'] = le.fit_transform(data['Embarked'])
data['Cabin'] = le.fit_transform(data['Cabin'])
data['Persons_accompanied'] = le.fit_transform(data['SibSp']+data['Parch'])
data['Title_str'] = data['Name'].str.split(',', expand=True)[1].str.split('.', expand=True)[0]
data['Title'] = le.fit_transform(data['Title_str'])
```

```python
X = data[['PassengerId', 'Age', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Cabin', 'Embarked', 'Fare']]
y = data['Survived']
```

```python
X_train, X_test,y_train, y_test = train_test_split(X,y,test_size=0.25,shuffle=True)
```

```python
#lsvc = LinearSVC(verbose=0)
#model = RandomForestClassifier()
```

```python
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 200, num = 10)]

max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]

max_features = ['auto', 'sqrt']

min_samples_split = [2, 5, 10]

min_samples_leaf = [1, 2, 4]
rf_grid = {'n_estimators': n_estimators,
           'max_features': max_features,
           'max_depth': max_depth,
           'min_samples_split': min_samples_split,
           'min_samples_leaf': min_samples_leaf}

model = RandomizedSearchCV(RandomForestClassifier(), rf_grid, scoring = 'accuracy', cv = 5,verbose=3)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
params = model.best_params_
print(params)
```
